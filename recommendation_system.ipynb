```python
import pandas as pd

data = {
    "Users" : ["A","A","A","B","B","B","C","C","C"],
    "Movie": ["Avengers","Batman","Spiderman",
              "Avengers","Batman","Superman",
              "Avengers","Spiderman","Superman"],
    "Rating": [5,4,5,5,3,4,4,5,5]
}

df = pd.DataFrame(data)
print(df)
```

Convert to User-Movie Matrix

```python
matrix = df.pivot_table(index="Users", columns="Movie", values="Rating")
print(matrix)
```

Calculate the Similarity Between Users 

```python
similarity = matrix.T.corr()
print(similarity)
```

Recommend Movie

Example: Recommend movie for User A

```python
user = "A"
similar_users = similarity[user].sort_values(ascending=False)
print(similar_users)
```

```python
recommended = matrix.loc["B"].dropna()
print("Recommended Movies:", recommended.index.tolist())
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(matrix.fillna(0))

print(similarity)
```
