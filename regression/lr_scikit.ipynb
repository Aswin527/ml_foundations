```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Dataset
'''X = np.array([[1], [2], [3], [4], [5]])   # Feature (Hours studied)
y = np.array([2, 4, 6, 8, 10])            # Target (Marks)'''

X = np.array([[2], [3], [4]])   # Hours studied
y = np.array([89, 90, 92])      # Marks

# Create model
model = LinearRegression()

# Train the model
model.fit(X, y)

# Predictions
y_pred = model.predict(X)

# Model parameters
print("Slope (weight):", model.coef_[0])
print("Intercept (bias):", model.intercept_)

# Visualization
plt.scatter(X, y, label="Actual Data")      # Data points
plt.plot(X, y_pred, label="Regression Line") # Best fit line

plt.xlabel("Hours Studied")
plt.ylabel("Marks")
plt.title("Linear Regression Visualization")

plt.legend()
plt.show()
```

```python
new_hours = np.array([[5.2]])
prediction = model.predict(new_hours)

print("Predicted Marks:", prediction[0])
```
